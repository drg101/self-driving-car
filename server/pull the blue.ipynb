Pulling the blue from a recycling bin.

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import pandas as pd
import os
```

```python
img_x = cv2.imread('data_recycle_bin/z.jpg')
img_y = cv2.imread('data_recycle_bin/y.jpg')
```

```python
def dispIm(im, label="pic"):
    plt.imshow(cv2.cvtColor(im,cv2.COLOR_BGR2RGB))
    plt.title(label)
    plt.show()

def dispGSIm(im, label='pic'):
    plt.imshow(im, cmap='gray', vmin = 0, vmax = 255,interpolation='none')
    plt.title(label)
    plt.show()
```

```python
dispIm(img_x)
dispIm(img_y)
```

```python
b = img_x[:,:,0]
g = img_x[:,:,1]
r = img_x[:,:,2]
```

```python
dispGSIm(b,'blue channel')
dispGSIm(g,'green channel')
dispGSIm(r,'red channel')
```

```python
bmr = cv2.subtract(b,cv2.multiply(1.3,r))
print(type(bmr))
print(b)
print('-')
print(r)
print("=")
print(bmr)
# bmr[bmr < 0] = 0
dispGSIm(bmr)
e,bm2rt = cv2.threshold(bmr, 40, 255, cv2.THRESH_BINARY)
dispGSIm(bm2rt)
```
